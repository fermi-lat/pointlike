#Fitting: optimizing the likelihood function
This notebook expands on the likelihood notebook by showing how fitting is implemented.
Fitting is managed by a subclass of <tt>BandLikeList</tt>, <a href="http://glast.stanford.edu/cgi-bin/cvsweb-SLAC/pointlike/python/uw/like2/views.py?view=markup"><tt>views.LikelihoodViews</tt></a>.  
<p>First we set up  ROI #840 as before, but instantiate the <tt>LikelihoodViews</tt> object instead using a convenience function.

```python
import os
from uw.like2 import views
os.chdir('/tmp/pointlike_demo')
%time likeviews = views.make_views(840)
print likeviews
```

This class implements a function, <tt>fitter_view</tt>, that returns a <i>view</i> of the likelihood function. The view is a function, returning the negative of the log likelihood for the scipy minimizing function.

```python
fv = likeviews.fitter_view()
print 'fitter_view function value: %.1f' % fv()
```

```python
%time fv.maximize();
likeviews.parameter_summary()
```

A useful view of the likelihood function is provided by the <tt>plot</tt> function, here looking at a single parameter:

```python
fv.plot(1)
```

The solid blue line is the log likelihood as a function of that parameter only. The dashed green like the parabolic expectation from the gradient and covariance matrix. It defines "sigma units" for the x-axis. The solid red line is the gradient, in sigma units, scale to the right. Finally the diamond is the current value. The blue curve is generally narrower than the parabola due to correlations.
<p>The <tt>plot_all</tt> function shows all variables, perhaps indicating some problems with the fit.

```python
fv.plot_all()
```

###Parameter subsets
The function <tt>fitter_view</tt> may define a view with a subset of the free parameters. This done with a source name, or a list of parameter indices. 

```python
source_name='P7R42722'
fv2 = likeviews.fitter_view(source_name)
fv2.parameter_summary()
fv2.maximize()
fv2.parameter_summary()
fv2.plot_all()
```

Finally, choose a view with only the normalization: the error is smallest since there are no other parameters, and so the measured likelihood matches the parabolic estimate.

```python
fv3 = likeviews.fitter_view([1])
fv3.maximize()
fv3.plot_all()
```

##Energy flux view
This is used to construct SED plots using the capability of the <tt>BandlikeList</tt> class to select a subset of bands for the likelihood evaluation sum. The function <tt>LikelihoodViews.energy_flux_view(source_name, energy)</tt> returns a function that is the likelihood as a function of the energy flux at the specified energy 

```python
with likeviews.energy_flux_view(source_name) as efv:
    x = np.linspace(18,24,21)
    y =np.array(map(efv,x ))
    fig,ax = plt.subplots(figsize=(3,3))
    ax.plot(x,y-y.max())
    plt.setp(ax, xscale='linear', ylim=(-10,0.1), xlabel='Energy flux (eV/cm**2/s)', ylabel='log likelihood')
    ax.grid()
    ax.set_title('%s flux' % source_name, size=12) 
fig
```

##TS map view
The function <tt>LikelihoodViews.tsmap_view(source_name)</tt> returns a function that evaluates the likelihood for the given source as a function of its position.

```python
s = likeviews.sources.find_source(source_name); ra,dec = s.skydir.ra(), s.skydir.dec()
with likeviews.tsmap_view(source_name) as mv:
    fig, ax = plt.subplots(figsize=(3,3))
    d=2
    delta = np.linspace(-d, d, 30)
    y = np.array([mv((ra, dec+x/60.)) for x in delta])
    ax.plot(delta, y)
    ax.axvline(0, color='k')
    plt.setp(ax, xlabel=r'$\Delta \mathrm{Dec\ (arcmin)}$', ylim=(-10,1), xlim=(-d, d), ylabel='log likelihood'); ax.grid()
    ax.set_xticks(np.linspace(-d,d,5))
    ax.set_title('%s localization' %source_name, size=12)
fig
```

Note the use of <tt>with</tt> in previouis two examples. All of the views support this. Since they all involve changing parameters
in the basic likelihood engine, this is a way to guarantee that the configuration not inadvertently changed, in this last case the position of the source.
