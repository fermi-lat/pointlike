There have been several studies of SNR spectra lately, which have had to estimate the effect of uncertaintiy in our IEM on the measured spectrum. I'm approaching the same question, but from an all-sky point of view, a straight-forward technique to apply to all sources.

##1 The pointlike binned likelihood function.
I start by reminding you about the likelihood function used by _pointlike_.

###1.1 Likelihood definition
Consider $K$ energy bands, each with $N_k$ angular bins in an ROI, with $M$ component sources. There are separate bands for front and back, the angular bin size is small compared with the PSF, and there are four energy bands per decade, 14 total from 100 MeV to 316 GeV. The bin sizes are small enough that the resolution is equivalent to unbinned likelihood, the limit for zero-sized bins.

Let:  
&ensp; $\ m_{kj}$: The number of measured events in energy band $k$ and angular bin $j$   
&ensp; $\ f_{ikj}$: the unnomalized exposure-weighted flux for source $i$ in band $k$, bin $j$     
&ensp; $\ F_{ik} = \sum\limits_j f_{ikj}$    
&ensp; $\ \alpha_i$: The prefactor, or multiplicative parameter for source $i$, to be determined. (There are also of course _shape_ parameters, not specified here.) Thus $\alpha_i f_{ijk}$ is the predicted counts from source $i$ for band  $k$, bin $j$. 

Using Poisson statistics, the maximum likelihood solution for the $\alpha$ parameters is the maximum with respect to $\alpha_i$ of the log likelihood,   

&ensp;&ensp; $$\ w(\alpha) = \sum\limits_k \[ \sum\limits_j\[m_{kj} \ln(\sum\limits_i \alpha_i f_{ikj}) \] -\sum\limits_i\alpha_i F_{ik} \] $$  

or the set of $M$ equations  

&ensp;&ensp;$  \frac{\partial w}{\partial \alpha_i} = 0 =\sum\limits_k \[ \sum\limits_j m_{kj} \( \frac{f_{ikj}}{\sum\limits_l\alpha_l f_{lkj}} \) - F_{ik} $

###1.2 Toy example and point-source resolution estimate
A simple, but relevant, "toy" example is a single point source with a constant background, for one energy band and conversion type. Center the ROI, with solid angle $\Delta \Omega$, on the source and assume azimuthal symmetry and that the PSF is contained in the ROI. Thus ignore the $j$ and $k$ indeces. For simplicity, fold the exposure, a constant, into the $\alpha$ definitions. Then, showing explicitly the dependence on the angle $\theta$ with respect to the point source

&ensp; $\ w(\alpha; \theta) = \ln(\alpha_0 P(\theta) + \alpha_1 ) -(\alpha_0  + \alpha_1 \Delta \Omega)$

Now $\alpha_0$ and $\alpha_1$ are respectively the number of photons from the source and the background photon density.

This form can be used to determine the expected resolution for the parameters. Assuming that the background dominates, the expected variances are:

&ensp; $\mathrm{Var}(\alpha_0) = \alpha_1 / \int P(\theta)^2 \mathrm{d} \Omega $

&ensp; $\mathrm{Var}(\alpha_1) = \alpha_1 / \Delta \Omega$

These are the number of background photons in the "footprint" of the point source, defined by the integral of the square of the PSF, and $\alpha_1$ divided by the total number of background photons.


###1.3 Caveats
This procedure assumes perfect knowledge of the model coefficients $f_{ikj}$. I want to deal with the special case of the IEM, or Galactic diffuse, for which there is an additional uncertainty in the spectral values; that is, a systematic uncertainty for each energy band. Since this involves an integral of the diffential flux over the range of energies, a factor of 1.8 in this case, the dispersion may also be a factor.

##4 Source consistency

We have another measure of the goodness-of-fit: the likelihood ratio for individual point sources.Consider an individual band $j$ and point source $i$ as defined above. Consider the log likelihood function $w_j(\alpha_i)$ with all the other parameters fixed by the overall fit. Then determine it with $\alpha_i$ set to the over-all fit value, and determine the value that maximize the likelood for this band only. This is the procedure that we use to make SED plots, for example:

```python
display(get_image('../uw22/sedfig/PSR_J2214p3000_sed.png', size=sed))
```

The curve represents the best fit to all the data, assuming the pulsar spectral model. Each data point is the measurement of the flux for that band (using both front and back), optimizing the likelihood as described. Then the goodness-of-fit parameter is the difference between the total likelihood, and a larger number, which is the combined likelihood for each band. We express likelihood ratios as a Test Statistic (TS) which is twice the log of the ratio. For a single bin, the expected distribution is chi squared with one degree of freedom. The plot above shows the signed square root of this quantity, the sign from the sign of the difference between the measured and predicted value for the flux. The distribution for this is just the normal distribution. The overall fit quality is then the sum of the squares of these pulls, like a chi squared for linear least squares. In this case, the value is 11.8. Assuming 10 degrees of freedom, the corresponding p-value is 29.6%, as displayed.

###5.1 Adding systematic errors

The likelihood function that we are using has no simple way to add uncertainties to the values of "nusiance" parameters like the galactic flux values; yet the fundamental assumption for maximizing the likelihood is that it is a faithful estimate of the probability of the experimental result, given the parameters. Thus we need to at least reduce the weight that the Poisson probability that we use gives to the numbers of events.

So look at the numbers: how many events are involved? I will use the relationship derived above, that the integral of the PSF squared defines an effective size

```python
display(get_image('plots/environment/psf_plot_uw27.png', size=(300,300)))
```

Now I look at an ROI along the galactic ridge, and measure the number of counts from the two main sources of diffuse emission in the size defined by the PSF.

```python
display(get_image('~/analysis/diffuse_systematics/counts_in_psf.png')) 
```

The red line is drawn at 1000 counts, for which the Poisson statistical error would be 3.2%. I treat this for now as a maximum value to allow for the systematic error associated with this correction, and will apply it to sources for which the counts is greater than this. So note that it would apply only to sources in the galactic plane, and only affect the Galactic diffuse below 1 GeV.

The above calculation uses the average diffuse in an ROI; it would be better to apply the local density, which I might do in the future.

###5.2 Unweighting the likelihood
The objective is to reduce the weights for the low-energy contributions to the likelihood, to account for our preliminary 3.2% systematic error for the value of the diffuse contribution. For the maximum likelihood technique, the maximum of the likelihood function is also the maximum of the log likelihood, the function $w$. At the maximum, the derivatives of $w$ with respect to all parameters must be zero, and the second derivatives define the covariance matrix. So I simply divide each term of the function $w$ by a factor which is the ratio of the number of background counts to 1000, or 1. This preserves the solution for that energy band, but lowers its weight with respect to the sum over all bands.

Note that this corresponds to assuming that the errors are uncorrelated, which is conservative.

The following plot shows the effect on our pulsar: note that the position of the first data point is the same, but that its error has increased, so that the pull is now 2, and the p-value is up to 36%.

##6 Summary

* Sources near the Galactic plane, except very bright ones like Vela, for energies below 1 GeV, are very sensitive to the IEM spectrum. It is important to take systematic uncertainties and the possible necessity for corrections, of this into account, an issue that becomes more important as the data set increases: the statistical error is now at or below 1% at 100 MeV, yet the systematic errors seem to be on the order of 3%.
   + The isotropic diffuse never reaches the 3% level; its systematic uncertainty is surely less, so there is no need to consider it
   + It is important to consider the systematics only below 1 GeV; above 10 GeV the IEM component is essentially irrelevant. From my point of view, the efforts to improve the spectral behavior of the IEM above a few GeV are not relevant.

* For all-sky catalog analysis I discuss here, there are large residuals for the counts distribution with ROI's near the plane, adjusting the IEM component resolves this.
* I introduced the concept of a goodness-of-fit parameter for sources; it indicates poor fits correlated with the IEM offsets, removing the offsets dramatically improves this as well
* I showed how to redefine the likelihood in an unbiased way, effectively reducing the number of events to unweight energy bands subject to IEM.
* It _matters_ for spectral fit parameters!
